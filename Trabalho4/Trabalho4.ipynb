```python
import pandas as pd

df_performance = pd.read_csv('/content/minkowski_performance.csv')
df_distancia = pd.read_csv('/content/distancia_poligonos.csv')
df_pontos = pd.read_csv('/content/pontos_log.csv')
```

```python
import matplotlib.pyplot as plt

df_performance['TotalVertices'] = df_performance['RobotVertexCount'] + df_performance['ObstacleVertexCount']

plt.figure(figsize=(10, 6))
plt.plot(df_performance['TotalVertices'], df_performance['ExecutionTimeMicroseconds'], marker='o')
plt.xlabel('Número Total de Vértices')
plt.ylabel('Tempo de Execução (microssegundos)')
plt.title('Desempenho do Algoritmo Minkowski Sum por Número de Vértices')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df_pontos['X'], df_pontos['Y'], s=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Comportamento dos Pontos ao Longo do Tempo')
plt.grid(True)
plt.show()
```
